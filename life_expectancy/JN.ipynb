```python
import pandas as pd
import os
import numpy as np
from typing import Type
#df = pd.read_csv("/Users/danielneves/00_DareData/090_pod_5/assignments/life_expectancy/data/eu_life_expectancy_raw.tsv", sep="\t")
```

```python
data_dir = os.path.join(os.getcwd(), 'data')
```

```python
df = pd.read_csv(os.path.join(data_dir, "eu_life_expectancy_raw.tsv"), sep="\t")
```

```python
df
```

```python
def split_column(df_: pd.DataFrame) -> pd.DataFrame:
    df = df_.copy()
    new_cols = df['unit,sex,age,geo\\time'].str.split(',', expand=True)
    new_cols.columns = ['unit', 'sex', 'age', 'region']
    df = pd.concat([df, new_cols], axis=1)
    df = df.drop(['unit,sex,age,geo\\time'], axis=1)
    return df
```

```python
df_splited = split_column(df)
```

```python
df_splited
```

```python
melted_df = df_splited.melt(id_vars=['unit', 'sex', 'age', 'region'], var_name='year', value_name='value')
melted_df
```

```python
melted_df['value'] = melted_df['value'].replace(to_replace=r'[^\d\.]+', value='', regex=True)

```

```python
not_null_values = melted_df[melted_df['value'].notnull()]
print(not_null_values['value'])
```

```python
melted_df['value'] = melted_df['value'].replace(to_replace='', value=np.nan)
```

```python
melted_df
```

```python
not_null_values = melted_df[melted_df['value'].notnull()]
print(not_null_values['value'])
```

```python
melted_df = melted_df.dropna(subset=['value'])
```

```python
not_null_values = melted_df[melted_df['value'].notnull()]
print(not_null_values['value'])
```

```python
melted_df['value'] = pd.to_numeric(melted_df['value'], errors='coerce')
```

```python
melted_df['value'] = melted_df['value'].astype(float)
```
